## PRIME NUMBERS


```python
number=int(input("enter a number:"))
if number > 1:
    for i in range(2,number):
        if number % i == 0:
            print(number,"is not a prime number")
            break
    else:
        print(number,"is a prime number")
else:
    print(number,"is not a prime number")
```

## Hospital Bill

```python
room_days=int(input("enter a days:"))
lab_tests=int(input("enter a tests:"))
room_charges=room_days*2000
lab_test_charges=lab_tests*300
doctor_fee=1500
medicine_charges=2400
total_bill=(room_charges+lab_test_charges+doctor_fee+medicine_charges)
if total_bill>10000:
    discount=total_bill * 0.10
else:
    discount=0
final_bill=total_bill-discount
print("room_cost:",room_charges)
print("lab_tests_cost:",lab_test_charges)
print("doctor_fee:",doctor_fee)
print("medicine_cost:",medicine_charges)
print("total_bill:",total_bill)
print("final_bill:",final_bill)
```
